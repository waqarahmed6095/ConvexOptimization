Find the **best-fit line** from the points **(1,-2),(2,3),(3,1)**. This is least square problem. 
The Solution for the problemLSPR.PNG


```
#Linear Algebra Solution
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import cvxpy as cp
np.set_printoptions(precision=4)

A=np.matrix('1,1; 1, 2; 1, 3')
b=np.matrix('-2;3;1')
alpha_linear_algebra=np.linalg.lstsq(A,b)[0]
print('The linear Algebra Solution:')
print(np.array(alpha_linear_algebra))
print('The value of the objective')

print(round(np.linalg.norm(A*alpha_linear_algebra-b),4))
print('.............................')

#Cvx Solution
[m,n]=np.shape(A)
alpha_cvx=cp.Variable(shape=(n,1))
objective=cp.Minimize(cp.norm2(A@alpha_cvx-b))
problem=cp.Problem(objective=objective)
solution=problem.solve()
print('The CVX solution: ')
print(alpha_cvx.value)
print('The value of the objective: ')
print(round(solution,4))
```
